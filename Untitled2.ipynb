```python
import numpy
import matplotlib.pyplot as plt
import pandas as pd

#import a csv file from google drive
from google.colab import drive
drive.mount('/content/drive')

#import Tweets-Barackobama.csv as csv
df = pd.read_csv('/content/drive/My Drive/Tweets-BarackObama.csv')



```

```python
# prompt: use regex to capture the characters before the first slash in every date column in an array

import re

def extract_year_from_date(date_string):
  """Extracts the year (characters before the first slash) from a date string."""
  match = re.match(r'([^/]+)', date_string)
  if match:
    return match.group(1)
  return None

# Apply the function to the 'Date' column and create a new 'Year' column
df['Year'] = df['Date'].apply(extract_year_from_date)

# Display the first few rows with the new 'Year' column
print(df[['Date', 'Year']].head())
```

```python
# checking the count of em dash present in the Tweet-text column and plot them by amount per year

import matplotlib.pyplot as plt
# Calculate the count of em dashes in each tweet
df['em_dash_count'] = df['Tweet-text'].str.count('—')

# Group by Year and sum the em dash counts
em_dash_counts_by_year = df.groupby('Year')['em_dash_count'].sum()

# Sort the years for better plotting
em_dash_counts_by_year = em_dash_counts_by_year.sort_index()

# Plot the results
plt.figure(figsize=(12, 6))
em_dash_counts_by_year.plot(kind='bar')
plt.title('Total Em Dash Count per Year in Tweet-text')
plt.xlabel('Year')
plt.ylabel('Total Em Dash Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Checking the count of En Dashes present in the Tweet-text field and plot them by year

import matplotlib.pyplot as plt
# Calculate the count of en dashes in each tweet
df['en_dash_count'] = df['Tweet-text'].str.count('–') # En dash is '–'

# Group by Year and sum the en dash counts
en_dash_counts_by_year = df.groupby('Year')['en_dash_count'].sum()

# Sort the years for better plotting
en_dash_counts_by_year = en_dash_counts_by_year.sort_index()

# Plot the results
plt.figure(figsize=(12, 6))
en_dash_counts_by_year.plot(kind='bar')
plt.title('Total En Dash Count per Year in Tweet-text')
plt.xlabel('Year')
plt.ylabel('Total En Dash Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
